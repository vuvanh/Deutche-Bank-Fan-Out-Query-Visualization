```python
import os
import asyncio
from tavily import AsyncTavilyClient

tavily_client = AsyncTavilyClient(api_key=os.getenv("API_KEY"))

async def research(query: str):
    response = await tavily_client.research(query)

    request_id = response.get("request_id")
    if not request_id:
        raise ValueError(f"No request_id in response: {response}")

    while True:
        result = await tavily_client.get_research(request_id)

        print("\n=== JSON RESPONSE ===")
        print(result)

        print("\n=== RESEARCH STATUS ===")
        status = result.get("status")
        print(status)

        if status == "completed":
            print(result.get("content"))
            print(result.get("sources"))
            return result

        if status == "failed":
            raise RuntimeError(f"Research failed: {result}")

        await asyncio.sleep(3)

result = await research(
    "Is it worth seeking to invalidate a euro-denominated mortgage in Poland?"
)
print(result)
```
